```python
# 1. تثبيت الأداة والمكتبات بشكل مستقر
!pip install -U demucs pydub soundfile > /dev/null
!apt-get install ffmpeg -y > /dev/null
print("✅ البيئة جاهزة ومستقرة تماماً.")

import os
import sys
import shutil
import subprocess
from pydub import AudioSegment
from google.colab import files
from demucs import separate

# إعداد المسارات
split_dir = "/content/chunks"
output_dir = "/content/output"
final_output = "/content/final_tracks"

# 🔍 البحث التلقائي عن الملف الصوتي المرفوع
supported_extensions = ('.mp3', '.wav', '.m4a', '.flac', '.ogg')
input_file = None
for file in os.listdir('/content'):
    if file.endswith(supported_extensions):
        input_file = os.path.join('/content', file)
        break

# تنظيف المجلدات المؤقتة
for d in [split_dir, output_dir, final_output]:
    if os.path.exists(d):
        shutil.rmtree(d)
    os.makedirs(d)

if input_file is None:
    print("❌ يرجى رفع ملف الصوت في القائمة الجانبية أولاً.")
else:
    print(f"🎵 تم العثور على الملف: {os.path.basename(input_file)}")
    print("📦 1. جاري تجزئة الملف العملاق عبر ffmpeg لضمان قراءة المدة كاملة...")

    # استخدام ffmpeg لتقسيم الملف بدقة إلى أجزاء مدتها 20 دقيقة (1200 ثانية)
    chunk_time = 1200
    output_pattern = os.path.join(split_dir, "chunk_%d.mp3")

    # أمر النظام للتقسيم الدقيق دون قراءة خاطئة للمدة
    cmd = ["ffmpeg", "-i", input_file, "-f", "segment", "-segment_time", str(chunk_time), "-c", "copy", output_pattern]
    subprocess.run(cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

    chunk_paths = [os.path.join(split_dir, f) for f in sorted(os.listdir(split_dir)) if f.startswith("chunk_")]

    print(f"✅ تم تقسيم الملف بنجاح إلى {len(chunk_paths)} أجزاء دقيقة.")
    print("\n🚀 2. جاري بدء المعالجة الصاروخية عبر الـ GPU السحابي...")

    for idx, path in enumerate(chunk_paths):
        print(f"⏳ جاري فصل الجزء رقم {idx + 1} من {len(chunk_paths)}...")
        sys.argv = ["demucs", "-n", "htdemucs_6s", "-o", output_dir, "--device", "cuda", path]
        try:
            separate.main()
        except Exception as e:
            print(f"⚠️ تنبيه في الجزء {idx + 1}: {str(e)}")

    print("\n🎵 3. جاري دمج وتجميع المسارات الستة النهائية تلقائياً وبجودة عالية...")
    instruments = ["vocals", "drums", "bass", "piano", "guitar", "other"]

    for inst in instruments:
        combined = AudioSegment.empty()
        for idx in range(len(chunk_paths)):
            # التأكد من الترتيب الصحيح للأجزاء أثناء الدمج
            track_chunk_dir = os.path.join(output_dir, "htdemucs_6s", f"chunk_{idx}")
            inst_path = os.path.join(track_chunk_dir, f"{inst}.wav")
            if os.path.exists(inst_path):
                combined += AudioSegment.from_wav(inst_path)

        if len(combined) > 0:
            combined.export(os.path.join(final_output, f"{inst}.mp3"), format="mp3", bitrate="320k")
            print(f"✅ تم تجميع مسار: {inst}.mp3 بنجاح.")

    print("\n🎉 مبروك! انتهت عملية الفصل بالكامل والملفات الستة متواجدة الآن في مجلد final_tracks.")
    print("🛟 يمكنك الآن النزول للخلية التالية بالأسفل، اختيار الآلات التي تريد دمجها، ثم تشغيلها مباشرة.")
```

```python
# كود دمج المسارات المفصولة حسب اختيارك
import os
from pydub import AudioSegment
from google.colab import files

# المسار الذي توجد به الملفات المفصولة حالياً
final_output = "/content/final_tracks"
merged_output_path = "/content/custom_merged_track.mp3"

# 🎵 1. حدد هنا ما تريد دمجه (اجعل الخيار True للدمج و False للاستبعاد)
choose_tracks = {
    "vocals":  True,   # الصوت البشري
    "piano":   False,  # البيانو
    "guitar":  False,  # الجيتار
    "bass":    True,   # الباص
    "drums":   True,   # الطبول
    "other":   False   # المؤثرات الأخرى
}

# التحقق من وجود مجلد الملفات
if not os.path.exists(final_output) or len(os.listdir(final_output)) == 0:
    print("❌ لم يتم العثور على المسارات المفصولة! تأكد من تشغيل كود الفصل بنجاح أولاً.")
else:
    mixed_audio = None
    tracks_to_mix = []

    # 2. قراءة الملفات المختارة فقط
    print("⏳ جاري تحضير المسارات المختارة للدمج...")
    for track_name, should_include in choose_tracks.items():
        if should_include:
            file_path = os.path.join(final_output, f"{track_name}.mp3")
            if os.path.exists(file_path):
                print(f"➕ تم إضافة: {track_name}.mp3")
                tracks_to_mix.append(AudioSegment.from_mp3(file_path))
            else:
                print(f"⚠️ تنبيه: لم يتم العثور على ملف {track_name}.mp3 في المجلد.")

    # 3. دمج وميكساج الأصوات فوق بعضها بالتزامن
    if len(tracks_to_mix) == 0:
        print("❌ لم تقم باختيار أي مسار لدمجه! يرجى تحويل خيار مسار واحد على الأقل إلى True.")
    else:
        print("\n🎛️ جاري عمل الميكساج والدمج الصوتي الآن...")
        mixed_audio = tracks_to_mix[0]
        for next_track in tracks_to_mix[1:]:
            # دمج الأصوات فوق بعضها بشكل متزامن تماماً (Overlay)
            mixed_audio = mixed_audio.overlay(next_track)

        # 4. تصدير الملف النهائي وتحميله
        print("💾 جاري حفظ وتصدير الملف النهائي بجودة 320k...")
        mixed_audio.export(merged_output_path, format="mp3", bitrate="320k")

        print("📥 جاري بدء التحميل التلقائي للملف المدمج...")
        files.download(merged_output_path)
        print("🎉 مبروك! اكتمل الدمج بنجاح وافحص مجلد التنزيلات بجهازك باسم (custom_merged_track.mp3).")
```
